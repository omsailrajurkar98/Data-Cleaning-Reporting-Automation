```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from datetime import datetime

file_path = "supermarket_salesmissing - Sheet1.csv"

df = pd.read_csv(file_path)

print("="*70)
print("DATA CLEANING & REPORTING AUTOMATION")
print("="*70)

print("\nDATASET SHAPE")
print(df.shape)

print("\nFIRST 5 ROWS")
display(df.head())

print("\nLAST 5 ROWS")
display(df.tail())

print("\nCOLUMN NAMES")
display(pd.DataFrame(df.columns, columns=["Columns"]))

print("\nDATA TYPES")
display(pd.DataFrame(df.dtypes, columns=["Data Type"]))

rows_before = len(df)
missing_values_before = df.isnull().sum().sum()
duplicates_before = df.duplicated().sum()

quality_before = pd.DataFrame({
    "Metric": [
        "Rows",
        "Columns",
        "Missing Values",
        "Duplicate Records"
    ],
    "Value": [
        len(df),
        len(df.columns),
        missing_values_before,
        duplicates_before
    ]
})

print("\nDATA QUALITY BEFORE CLEANING")
display(quality_before)

missing_report = pd.DataFrame({
    "Column": df.columns,
    "Missing Values": df.isnull().sum().values
})

missing_report = missing_report.sort_values(
    by="Missing Values",
    ascending=False
)

print("\nMISSING VALUE REPORT")
display(missing_report)

missing_report.plot(
    x="Column",
    y="Missing Values",
    kind="bar",
    legend=False,
    figsize=(12,6)
)

plt.title("Missing Values Per Column")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

print("\nDUPLICATE RECORDS FOUND:", duplicates_before)

df.drop_duplicates(inplace=True)

for col in df.columns:

    if df[col].dtype == "object":

        mode_val = df[col].mode()

        if len(mode_val) > 0:
            df[col] = df[col].fillna(mode_val[0])

    else:

        median_val = df[col].median()

        df[col] = df[col].fillna(median_val)

df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

print("\nCLEANED DATASET PREVIEW")
display(df.head())

quality_after = pd.DataFrame({
    "Metric": [
        "Rows",
        "Columns",
        "Remaining Missing Values",
        "Remaining Duplicates"
    ],
    "Value": [
        len(df),
        len(df.columns),
        df.isnull().sum().sum(),
        df.duplicated().sum()
    ]
})

print("\nDATA QUALITY AFTER CLEANING")
display(quality_after)

print("\nSUMMARY STATISTICS")
display(df.describe(include="all"))

numeric_cols = df.select_dtypes(
    include=["int64", "float64"]
).columns

categorical_cols = df.select_dtypes(
    include=["object"]
).columns

if len(numeric_cols) > 1:

    print("\nCORRELATION HEATMAP")

    corr = df[numeric_cols].corr()

    plt.figure(figsize=(10,8))

    plt.imshow(corr)

    plt.colorbar()

    plt.xticks(
        range(len(corr.columns)),
        corr.columns,
        rotation=90
    )

    plt.yticks(
        range(len(corr.columns)),
        corr.columns
    )

    plt.title("Correlation Heatmap")

    plt.show()

if len(numeric_cols) > 0:

    print("\nDISTRIBUTION ANALYSIS")

    plt.figure(figsize=(8,6))

    df[numeric_cols[0]].hist(
        bins=30
    )

    plt.title(
        f"Distribution of {numeric_cols[0]}"
    )

    plt.xlabel(
        numeric_cols[0]
    )

    plt.ylabel(
        "Frequency"
    )

    plt.show()

if len(numeric_cols) > 0:

    print("\nOUTLIER ANALYSIS")

    plt.figure(figsize=(8,6))

    plt.boxplot(
        df[numeric_cols[0]]
    )

    plt.title(
        f"Outlier Detection: {numeric_cols[0]}"
    )

    plt.show()

if len(categorical_cols) > 0:

    category = categorical_cols[0]

    top_categories = (
        df[category]
        .value_counts()
        .head(10)
    )

    print("\nTOP CATEGORY ANALYSIS")

    display(top_categories)

    plt.figure(figsize=(10,6))

    top_categories.plot(
        kind="bar"
    )

    plt.title(
        f"Top 10 Categories - {category}"
    )

    plt.ylabel(
        "Count"
    )

    plt.xticks(
        rotation=45
    )

    plt.show()

dtype_counts = (
    df.dtypes
    .astype(str)
    .value_counts()
)

print("\nDATA TYPE DISTRIBUTION")

display(dtype_counts)

plt.figure(figsize=(8,8))

plt.pie(
    dtype_counts.values,
    labels=dtype_counts.index,
    autopct="%1.1f%%"
)

plt.title(
    "Dataset Data Types"
)

plt.show()

report = pd.DataFrame({

    "Metric": [

        "Rows Before Cleaning",
        "Rows After Cleaning",
        "Columns",
        "Duplicates Removed",
        "Missing Values Before",
        "Missing Values After"

    ],

    "Value": [

        rows_before,
        len(df),
        len(df.columns),
        duplicates_before,
        missing_values_before,
        df.isnull().sum().sum()

    ]
})

print("\nFINAL AUTOMATED REPORT")
display(report)

print("\n" + "="*70)
print("PROJECT SUMMARY")
print("="*70)

print(f"Execution Time : {datetime.now()}")
print(f"Rows Processed : {rows_before}")
print(f"Rows Remaining : {len(df)}")
print(f"Duplicates Removed : {duplicates_before}")
print(f"Missing Values Fixed : {missing_values_before}")
print("\nAutomation Completed Successfully")
```
